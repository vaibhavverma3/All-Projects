#Handling Null Values - Data cleaning
Taking care of None and NaN values

```python
df.fillna({"director":"Unknown Director","cast":"Unknown Cast","country":"Unknown Country",'rating':'Unknown Rating'} , inplace=True)
df
```

```python
df.dropna(inplace=True)
#dropping remaining rows that have null values
```

```python
df.isnull().sum()
#all null values removed
```

#Unnesting the data

```python
# Checking which columns have nested values
# Initialize a list to store columns with nested values
nested_columns = []

# Iterate through each column and check for nested values
for col in df.columns:
    # Check if any cell in the column contains commas
    if df[col].apply(lambda x: isinstance(x, str) and ',' in x).any():
        nested_columns.append(col)

# Print the columns with nested values
print("Columns with nested values separated by commas:")
print(nested_columns)
```

```python
# Now we need to unnest columns like director, cast, country
cast_df=df[['title','cast']]
cast_df['list of cast']=cast_df['cast'].apply(lambda x: str(x).split(', '))
cast_df=cast_df.explode('list of cast')
cast_df.drop('cast',axis=1,inplace=True)
cast_df
```

```python
#unnesting country columns
country_df = df[["title", "country"]]
country_df["list_of_country"]=country_df.country.apply(lambda x: str(x).split(", "))
country_df = country_df.explode("list_of_country")
country_df.drop('country',axis=1,inplace=True)
country_df
```

```python
#unnesting director column
director_df = df[["title", "director"]]
director_df["list_of_director"]=director_df.director.apply(lambda x: str(x).split(", "))
director_df = director_df.explode("list_of_director")
director_df.drop('director',axis=1,inplace=True)
director_df
```

```python
# Merge df and country_df on 'Title'
merged_df = pd.merge(df, country_df, on='title', how='inner')

# Merge cast_df with the merged_df on 'Title'
merged_df = pd.merge(merged_df, cast_df, on='title', how='inner')

# Merge director_df with the merged_df on 'Title'
merged_df = pd.merge(merged_df, director_df, on='title', how='inner')
merged_df
```

# Correcting the date format in the of the date_added column

```python
# Strip leading and trailing spaces from the 'Date_added' column
df['date_added'] = df['date_added'].str.strip()

# Convert 'Date_added' to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y')
```

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month_name()
```

#3.  What is the best time to launch a TV show/Movie?
  - Q3 a Find which is the best week to release the Tv-show or the movie. Do the analysis separately for Tv-shows and Movies

```python
# Extract the week number and create a new column 'Week'
df['Week'] = df['date_added'].dt.isocalendar().week
# Filter DataFrame for movies
movies_df = df[df['type'] == 'Movie']

# Group by week and count the total number of movies and TV shows for each week
movies_by_week = movies_df.groupby('Week').size().sort_values(ascending=False)
print(movies_by_week)
```

Insight: From above data we can say most number of movies are released in the first (1st) week of the year

```python
tv_shows_df = df[df['type'] == 'TV Show']
tv_shows_by_week = tv_shows_df.groupby('Week').size().sort_values(ascending=False)
print(tv_shows_by_week)
```

Insight: From above data, we can say the most number of TV shows are added in the 27th week

- Q3 b Find which is the best month to release the Tv-show or the movie. Do the analysis separately for Tv-shows and Movies

```python
# Extract month from 'Date_added' column
df['month_added'] = df['date_added'].dt.month_name()

# Filter data for TV shows and movies separately
movies_df = df[df['type'] == 'Movie']
# Group by week and count the total number of movies and TV shows for each week
movies_by_month = movies_df.groupby('month_added').title.nunique().sort_values(ascending=False)
print(movies_by_month)
```

```python
# Filter data for movies separately
movies = df[df['type'] == 'Movie']

# Group by month and count the number of movies added in each month
movies_by_month = movies.groupby('month_added').size().reset_index(name='Count')

# Plottingm ovies separately
#making a barplot for the above data of Monthwise Movies
#count of Movies in y axis
#Month in x axis
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
sns.barplot(data=movies_by_month, x='month_added', y='Count',hue='month_added')
plt.title('Number of Movies Released by Month')
plt.xlabel('Month')
plt.ylabel('Number of Movies Released')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Insight: From above graph and data we can say that most movies were added in the month of July (565 Movies)

```python
#now repeating the above process for TV shows
#Filter data for TV shows
tv_show_df = df[df['type'] == 'TV Show']
tv_shows_by_month = tv_shows_df.groupby('month_added').title.nunique().sort_values(ascending=False) #grouping tv shows by month added and counting TV show titles
print(tv_shows_by_month)
```

```python
#making a barplot for the above data of monthwise TV shows
#count of TV shows in y axis
#month in x axis
tv_shows = df[df['type'] == 'TV Show']
tv_shows_by_month = tv_shows.groupby('month_added').size().reset_index(name='Count')

plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
sns.barplot(data=tv_shows_by_month, x='month_added', y='Count', hue='month_added')
plt.title('Number of TV Shows Released by Month')
plt.xlabel('Month')
plt.ylabel('Number of TV Shows Released')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#6. Find after how many days the movie will be added to Netflix after the release of the movie (you can consider the recent past data)
Hint : We want you to get the difference between the columns having date added information and release year information and get the mode of difference. This will give an insight into what will be the better time to add in Netflix

```python
# Convert 'release_year' to datetime format
df['release_year'] = pd.to_datetime(df['release_year'], format='%Y')

# Calculate the difference between 'date_added' and 'release_year'
df['days_to_netflix'] = (df['date_added'] - df['release_year']).dt.days

# Find the mode of 'days_to_netflix'
mode_days = df['days_to_netflix'].mode()[0]

# Print the mode of 'days_to_netflix'
print(f"After how many days the movie is advisable to be added to netflix after its release : {mode_days} days")
```
